As you can see, the responses yield unorganized data.  We need to clean the data above to allow data analysis.

```python
import csv

## Below code takes courses from Data Science MPS catalog (taken from UMBC website) and creates dictionary with 'course_field_number': 'course_description'.

course_dict = {}

# Open catalog for reading.
with open('UMBC_MPS_DS_Course_Catalog.txt', 'r') as f:

# Create dict to store course code and name.
    global course_dict

    for line in f:
        begin_index = line.find('*')

        if begin_index == -1:
          continue

        elif line[4:8].strip('- ').upper() != line[4:8].strip('- '):
            continue

        elif line == '':
            continue

        course_dict[line[4:12].strip('- ')] = line[13:].strip('<#>\n').strip('- ').strip('"')

# Write dict to csv file for later reading.
with open('course_list.csv', 'w') as csvfile:

    writer = csv.writer(csvfile)
    writer.writerow(['Course', 'Description'])

    for course in course_dict:
      writer.writerow([course, course_dict[course]])
```

Now, we will clean the data by using the dictionary stored in 'course_list.csv'.

---



```python
course_count = {}

for response_raw in list(df["12. Other courses taking this semester"]):
  # Removes unwanted punctuation in response, lowercases all words, and puts commas in natural spots.
  response = str(response_raw)
  response_split = response.replace(':', ',').replace('|', ',').replace(';', ',').replace('and', ',').replace('-', ',').lower().split(',')

  # After creating a list of 'keywords' from the responses,
  # below checks to see if responses contain courses from catalog and counts it.
  for segment in response_split:
    segment_fixed = segment.strip()

    for course in course_dict:
      if course.lower() == segment_fixed or course_dict[course].lower() == segment_fixed:
        if course not in course_count:
          course_count[course] = 1
        else:
          course_count[course] += 1

# Below writes course_count dict to a csv.
with open('course_count.csv', 'w') as csvfile:
  writer = csv.writer(csvfile)
  writer.writerow(['Course', 'Count'])

  for course in course_count:
    writer.writerow([course, course_count[course]])



```

* Now, we have a csv file with all the courses mentioned and how many times.  
* We are ready to make a pie chart with this data.

```python
# Reads csv file made in code above.
df1 = pd.read_csv('course_count.csv')

# Finally, makes a pie chart with data.
px.pie(df1, names='Course', values='Count', title='Courses Taken with DATA690')
```

Solution of Bonus Question using Pandas

```python
list_answers = list(df["12. Other courses taking this semester"].dropna())
list_answers
```

```python
list_answers.pop(3)
list_answers
```

```python
list_answers.pop(11)
list_answers
```

```python
list_courses = []

for answer in list_answers:
    if "," in answer:
        courses = answer.split(",")
        for course in courses:
            list_courses.append(course)

    elif ";" in answer:
        courses = answer.split(";")
        for course in courses:
            list_courses.append(course)

    elif ("and" in answer) and (answer != 'Deep Learning and LLM'):
        courses = answer.split("and")
        for course in courses:
            list_courses.append(course)
    else:
        list_courses.append(answer)

list_courses
```
