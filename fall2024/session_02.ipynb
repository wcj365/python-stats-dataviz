## Step 2 - Explore Student Age

```python
yob_list = list(df["8. My Birth Year"])
yob_list
```

```python
cleansed_yob_list = []  # Empty list
cleansed_yob_list       # the same as print(cleansed_yob)
```

```python
for yob in yob_list:
    cleansed_yob = yob[-4:]
    cleansed_yob_list.append(cleansed_yob)

cleansed_yob_list
```

```python
age_list = []

for yob in cleansed_yob_list:
    age = 2024 - int(yob)
    age_list.append(age)

age_list
```
