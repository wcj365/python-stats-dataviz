```python
# <11>
# Now let's look at the tuition growth rate year over year
# We need to calculate UMBC tuition change percentage each year
# First convert the TUITIONFEE_IN column to a Python List

tuition_list = list(umbc["TUITIONFEE_IN"])
tuition_list
```

```python
# Loop through the list and calculate the % change each year over the prior year

pct_change = []
prev_year_tuition = np.NaN
for tuition in tuition_list:
    pct_change.append((tuition - prev_year_tuition) * 100 / prev_year_tuition)
    prev_year_tuition = tuition
pct_change   
```

```python
# <12>
# Since we want to do the same calculation for JHU. 
# Instead of doing it piecemeal as we did for UMBC
# Let's create a function which can be reused for any college.
# This function takes a list of tuitions and return a list of percentage changes

#(Write code here)

def calculate_percentage(tuition_list):
    pct_change = []
    prev_year_tuition = np.NaN
    for tuition in tuition_list:
        pct_change.append((tuition - prev_year_tuition) * 100 / prev_year_tuition)
        prev_year_tuition = tuition
    pct_change = [round(x, 2) for x in pct_change]  
    return pct_change   
```

```python
# <15>
# Let's calculate the tuition percentge growth using the function defined earlier

#(Write code here)

jhu_tuition_list = list(jhu["TUITIONFEE_IN"])
jhu_pct_change = calculate_percentage(jhu_tuition_list)
jhu_pct_change
```
