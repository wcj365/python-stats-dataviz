```python
import pandas as pd

print("Pandas Version is ", pd.__version__)

import plotly 

print("Plotly Version is", plotly.__version__)

import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv("/content/wdi_data.csv")
df.head(10)
```

```python
# df.drop(columns=["Unnamed: 0"], inplace=True)
df_new = df.drop(columns=["Unnamed: 0"])
df_new.head()
```

```python
df.head()
```

```python
# df_ind = df_new.query("'Country Code' == 'IND'")  NOT WORKING b/c the space between the column name.

df_ind = df_new[df_new["Country Code"] == 'IND']
df_ind
```

```python
df_ind_gdppcap = df_ind.query("indicator == 'NY.GDP.PCAP.PP.CD'")

df_ind_gdppcap
```

```python
df_ind_gdppcap.info()
```

```python
fig = go.Figure()

trace_0 = go.Bar(      
    x=df_ind_gdppcap["Year"],    
    y=df_ind_gdppcap["value"]
)

fig.add_trace(trace_0)

fig.update_layout(
    title="India GDP per Capita Over Time",
    xaxis_title="YEAR",
    yaxis_title="GDP per Capita"
)

fig.show()
```

```python
fig = px.bar(
    data_frame=df_ind_gdppcap,
    title="India GDP per Capita Over Time",
    labels={"value":"GDP per Capita", "Year":"YEAR"},
    orientation="v",
    x="Year",
    y="value"
)

fig.update_xaxes(tickangle=45)

fig.show()
```
