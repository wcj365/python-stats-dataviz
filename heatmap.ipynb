# UCI - Heart Disease Dataset

- age: age of the person in years.
- sex: 1 for male 0 for female
- cp: chest pain (0,1,2,3)
- trestbps: rest BP
- chol: cholestoral
- fbs: blood sugar on fasting.
- restecg: electrocardiographic
- thalach: maximum heart rate.
- exang: exercise with angina
- oldpeak: heart condition while exercising.
- slope: slope of the heart while exercising
- ca: indicate the blood movement
- thal: thalium stress (the more the danger)
- num: 1-person tent to have disease, 0-person dosnt tent to have any heart disease.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

#data_url = "https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/data/heart_disease_uci.csv"
data_url = "https://raw.githubusercontent.com/wcj365/python-stats-dataviz/master/data/cleveland_heart_disease.csv"
```

```python
df = pd.read_csv(data_url)

df.shape
```

```python
df.sample(5)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot=True, cmap=("RdYlGn"))
plt.show()
```
