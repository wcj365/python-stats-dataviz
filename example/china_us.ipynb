```python
import pandas as pd
import plotly.express as px
```

```python
df_imp = pd.read_excel("https://github.com/wcj365/python-stats-dataviz/raw/refs/heads/master/example/Imports_from_Counterpart_Countries.xlsx", skiprows=6)
df_imp.drop(columns=["Unnamed: 0"], inplace=True)
df_imp.rename(columns={"Unnamed: 1" : "Country"}, inplace=True)
df_imp.head()
```

```python
df_imp =pd.melt(df_imp, id_vars=["Country"], var_name='Year', value_name='Import')
df_imp
```

```python
df_exp = pd.read_excel("https://github.com/wcj365/python-stats-dataviz/raw/refs/heads/master/example/Exports_to_Counterpart_Countries.xlsx", skiprows=6)
df_exp.drop(columns=["Unnamed: 0"], inplace=True)
df_exp.rename(columns={"Unnamed: 1" : "Country"}, inplace=True)
df_exp.head()
```

```python
df_exp =pd.melt(df_exp, id_vars=["Country"], var_name='Year', value_name='Export')
df_exp
```

```python
df = df_imp.merge(df_exp, on=["Country", "Year"], how="inner")
df
```

```python
df_us = df[df["Country"] == "United States"]
df_us
```

```python
df_us["Surplus"] = df_us["Export"] - df_us["Import"]
df_us
```

```python
fig = px.bar(df_us, x="Year", y="Export")
fig.show()
```

```python
fig = px.bar(df_us, x="Year", y="Import")
fig.show()
```

```python
fig = px.bar(df_us, x="Year", y="Surplus")
fig.show()
```
