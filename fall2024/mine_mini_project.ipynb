## Step 4 - Categorize Each Country

Create a new column "Group". If a country is one of the seven G7 countried, the value of "Group" would be "G7", if a country of one of the five BRICS countries, the value of "Group" would be "BRICS". Otherwise, the value of "Group" would be "REST of WORLD".

In this step, we categorize each country into one of the following groups:

- **G7**: For countries that are part of the G7 group.
- **BRICS - Original**: For the five founding BRICS countries.
- **BRICS - All Members**: For the expanded list of BRICS countries, including new members.
- **Rest of World**: For countries not in G7 or BRICS.

We define a function to assign each country to the appropriate group based on its country code and the specified version of BRICS membership.


```python
# Define G7 and BRICS groups based on country codes
G7_COUNTRIES = ["USA", "CAN", "GBR", "FRA", "DEU", "ITA", "JPN"]
BRICS_ORIGINAL = ["BRA", "RUS", "IND", "CHN", "ZAF"]
BRICS_ALL_MEMBERS = ["BRA", "RUS", "IND", "CHN", "EGY", "ETH", "IRN", "SAU", "ZAF"]

# Function to assign each country to its appropriate group
def assign_group(row, brics_version="original"):
    """
    Categorizes each country as 'G7', 'BRICS - Original', 'BRICS - All Members', or 'REST OF WORLD'
    based on its country code.

    Parameters:
    - row: DataFrame row containing country information.
    - brics_version (str): Defines which BRICS grouping to use, options are 'original' or 'all_members'.

    Returns:
    - str: The group name for the country.
    """
    country_code = row['Country Code'].strip().upper()  # Ensure country code is standardized
    if country_code in G7_COUNTRIES:
        return "G7"
    elif country_code in (BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL):
        return "BRICS - All Members" if brics_version == "all_members" else "BRICS - Original"
    else:
        return "REST OF WORLD"

# Apply the function twice for each BRICS version
data['Group_Original'] = data.apply(assign_group, axis=1, brics_version="original")
data['Group_All_Members'] = data.apply(assign_group, axis=1, brics_version="all_members")

# Display each group with its countries for both versions
print("G7 Countries:")
display(data[data['Group_Original'] == "G7"][['Country Name', 'Country Code', 'Group_Original']].drop_duplicates())

print("\nBRICS - Original Countries:")
display(data[data['Group_Original'] == "BRICS - Original"][['Country Name', 'Country Code', 'Group_Original']].drop_duplicates())

print("\nBRICS - All Members Countries:")
display(data[data['Group_All_Members'] == "BRICS - All Members"][['Country Name', 'Country Code', 'Group_All_Members']].drop_duplicates())

print("\nREST OF WORLD Countries:")
display(data[data['Group_Original'] == "REST OF WORLD"][['Country Name', 'Country Code', 'Group_Original']].drop_duplicates())


```

## Step 5: Aggregate Data

This step involves aggregating GDP and GDP Per Capita data at the group level (G7, BRICS, and Rest of World) for each year.

- For GDP, we calculate the **sum** for each group.
- For GDP Per Capita, we calculate the **mean** for each group.

We use a reusable function to apply these aggregations and display the results for both "BRICS - Original" and "BRICS - All Members" groupings.


```python
# Define a reusable function to aggregate and display GDP and GDP Per Capita data
def aggregate_and_display_data(brics_version="original"):
    """
    Aggregates and displays GDP and GDP Per Capita data for each group (G7, BRICS, and REST OF WORLD).

    Parameters:
    - brics_version (str): Specifies whether to aggregate data for "BRICS - Original" or "BRICS - All Members".

    Aggregated metrics:
    - GDP: Sum for each group.
    - GDP Per Capita: Mean for each group.
    """

    # Assign each country to its appropriate group (G7, BRICS, or REST OF WORLD)
    data['Group'] = data.apply(assign_group, axis=1, brics_version=brics_version)

    # Group the data by year and group, then aggregate GDP and GDP Per Capita
    aggregated_data = data.groupby(['Year', 'Group']).agg({
        'GDP ($Billions)': 'sum',
        'GDP Per Capita ($)': 'mean'
    }).reset_index()

    # Format GDP and GDP Per Capita values for readability
    aggregated_data['GDP ($Billions)'] = aggregated_data['GDP ($Billions)'].apply(lambda x: f"{x:,.0f}")
    aggregated_data['GDP Per Capita ($)'] = aggregated_data['GDP Per Capita ($)'].apply(lambda x: f"{x:,.0f}")

    # Display the aggregated data for the selected BRICS version
    selected_brics_label = "BRICS - All Members" if brics_version == "all_members" else "BRICS - Original"
    print(f"Aggregated Data for G7, {selected_brics_label}, and REST OF WORLD:")
    display(aggregated_data)

# Call the function for both BRICS versions
aggregate_and_display_data(brics_version="original")
aggregate_and_display_data(brics_version="all_members")

```

For each year, calculate the percentage share of each group's GDP. Check your calculation with the reference site.

## Step 6: Calculate Percentage Share of GDP

In this step, we calculate the percentage share of GDP for each group (G7, BRICS, and Rest of World) by year. We:

- Aggregate GDP values by group and year.
- Calculate each groups GDP as a percentage of the total GDP for that year.
- Display the GDP percentage share for both BRICS versions ("Original" and "All Members").

This is done using a reusable function for flexibility and clarity.


```python
# Define a reusable function to calculate and display the GDP percentage share for each group
def calculate_and_display_gdp_percentage_share():
    """
    Calculates and displays the GDP percentage share for G7, BRICS (both Original and All Members),
    and REST OF WORLD for each year. The results are displayed in a combined DataFrame for comparison.
    """
    all_data = []  # To store results for both BRICS versions

    # Loop through both BRICS versions to calculate and display the GDP percentage share
    for brics_version in ["original", "all_members"]:
        # Assign each country to its appropriate group for the current BRICS version
        data['Group'] = data.apply(assign_group, axis=1, brics_version=brics_version)

        # Aggregate GDP by year and group
        aggregated_data = data.groupby(['Year', 'Group']).agg({
            'GDP ($Billions)': 'sum'
        }).reset_index()

        # Calculate total GDP per year and compute percentage share for each group
        total_gdp_per_year = aggregated_data.groupby('Year')['GDP ($Billions)'].transform('sum')
        aggregated_data['GDP Percentage Share (%)'] = (aggregated_data['GDP ($Billions)'] / total_gdp_per_year) * 100
        aggregated_data['GDP Percentage Share (%)'] = aggregated_data['GDP Percentage Share (%)'].round(1)  # Round to 1 decimal

        # Add a column to indicate the BRICS version used in this calculation
        brics_label = "BRICS - All Members" if brics_version == "all_members" else "BRICS - Original"
        aggregated_data['BRICS Version'] = brics_label

        # Append the result to the list for combining later
        all_data.append(aggregated_data)

    # Combine results from both BRICS versions for display
    combined_data = pd.concat(all_data).reset_index(drop=True)

    # Display the GDP percentage share for each group
    print("GDP Percentage Share for BRICS (Original and All Members), G7, and REST OF WORLD:")
    display(combined_data)

# Call the function to calculate and display the GDP percentage share
calculate_and_display_gdp_percentage_share()


```

## Step 7 - Plot the GDP Percentage Share Over Time

## Step 7: Plot the GDP Percentage Share Over Time

In this step, we visualize the GDP percentage share over time for each group: G7, BRICS, and Rest of World. We use a stacked bar chart, with different colors representing each group. The function `plot_gdp_percentage_share_custom` can be used to plot either "BRICS - Original" or "BRICS - All Members" by adjusting the `brics_version` parameter.


```python
# Define a reusable function to plot GDP percentage share over time
def plot_gdp_percentage_share_custom(brics_version="original"):
    """
    Plots the GDP percentage share over time for each group (G7, BRICS, and Rest of World)
    using a stacked bar chart.

    Parameters:
    - brics_version (str): Specifies the BRICS version to use ("original" or "all_members").
    """
    # Label the BRICS version based on the parameter
    brics_label = "BRICS - All Members" if brics_version == "all_members" else "BRICS - Original"

    # Assign each country to its appropriate group for the specified BRICS version
    data['Group'] = data.apply(assign_group, axis=1, brics_version=brics_version)

    # Aggregate GDP by year and group, then calculate the GDP percentage share
    aggregated_data = data.groupby(['Year', 'Group']).agg({
        'GDP ($Billions)': 'sum'
    }).reset_index()

    total_gdp_per_year = aggregated_data.groupby('Year')['GDP ($Billions)'].transform('sum')
    aggregated_data['GDP Percentage Share (%)'] = (aggregated_data['GDP ($Billions)'] / total_gdp_per_year) * 100
    aggregated_data['GDP Percentage Share (%)'] = aggregated_data['GDP Percentage Share (%)'].round(1)

    # Pivot the data for easier plotting in stacked bar format
    pivot_data = aggregated_data.pivot(index='Year', columns='Group', values='GDP Percentage Share (%)').fillna(0)

    # Define custom colors for each group
    colors = {
        brics_label: '#ff5733',  # Red for BRICS
        'G7': '#4b4bff',          # Blue for G7
        'REST OF WORLD': '#32CD32' # Green for Rest of World
    }

    # Define the order in which groups will appear in the stacked bar chart
    group_order = [brics_label, 'G7', 'REST OF WORLD']

    # Create the figure
    fig = go.Figure()

    # Add each group as a trace in the stacked bar chart
    for group in group_order:
        if group in pivot_data.columns:
            fig.add_trace(go.Bar(
                x=pivot_data.index,
                y=pivot_data[group],
                name=group,
                marker_color=colors[group],
                text=pivot_data[group].apply(lambda x: f"{x:.1f}%"),  # Add percentage text
                textposition='inside'
            ))

    # Update the layout for the stacked bar chart
    fig.update_layout(
        title="Percentage Share of GDP Over Time",
        xaxis=dict(
            title="Year"
        ),
        yaxis=dict(title="GDP (%)"),
        barmode="stack",  # Stack bars for each year
        legend=dict(
            orientation="h",
            yanchor="top",
            y=1.1,
            xanchor="center",
            x=0.5
        )
    )

    # Show the figure
    fig.show()

# Call the function to plot GDP percentage share for both BRICS versions
plot_gdp_percentage_share_custom(brics_version="original")
plot_gdp_percentage_share_custom(brics_version="all_members")

```

## Step 8 - Plot the Total GDP Over Time

## Step 8: Plot the Total GDP Over Time

In this step, we visualize the total GDP for each group (G7, BRICS, and Rest of World) over time using a stacked bar chart. The function `plot_total_gdp_bar_chart_with_labels` allows plotting for either "BRICS - Original" or "BRICS - All Members" by specifying the `brics_version` parameter.


```python
# Define a reusable function to plot total GDP over time with labels
def plot_total_gdp_bar_chart_with_labels(brics_version="original"):
    """
    Plots the total GDP over time for each group (G7, BRICS, and Rest of World) using a stacked bar chart.

    Parameters:
    - brics_version (str): Specifies which BRICS grouping to use ("original" or "all_members").
    """
    # Determine BRICS label based on the parameter
    brics_label = "BRICS - All Members" if brics_version == "all_members" else "BRICS - Original"

    # Assign each country to its appropriate group based on the specified BRICS version
    data['Group'] = data.apply(assign_group, axis=1, brics_version=brics_version)

    # Aggregate total GDP by year and group
    aggregated_data = data.groupby(['Year', 'Group']).agg({
        'GDP ($Billions)': 'sum'
    }).reset_index()

    # Pivot the data to facilitate stacking in a bar chart
    pivot_data = aggregated_data.pivot(index='Year', columns='Group', values='GDP ($Billions)').fillna(0)

    # Define custom colors for each group
    colors = {
        brics_label: '#ff5733',  # Red for BRICS
        'G7': '#4b4bff',         # Blue for G7
        'REST OF WORLD': '#32CD32'  # Green for Rest of World
    }

    # Define the order of groups in the stacked bar chart
    group_order = [brics_label, 'G7', 'REST OF WORLD']

    # Create the figure
    fig = go.Figure()

    # Add a trace for each group, with labels showing GDP values
    for group in group_order:
        if group in pivot_data.columns:
            fig.add_trace(go.Bar(
                x=pivot_data.index,
                y=pivot_data[group],
                name=group,
                marker_color=colors[group],
                text=pivot_data[group].apply(lambda x: f"${x:,.0f}"),  # Format GDP values
                textposition='inside',
                insidetextanchor="middle"
            ))

    # Update layout for the stacked bar chart
    fig.update_layout(
        paper_bgcolor='white',
        plot_bgcolor='white',
        title="Total GDP Over Time",
        xaxis=dict(
            title="Year"
        ),
        yaxis=dict(
            title="GDP ($Billions)"
        ),
        barmode="stack",  # Stack bars by year
        legend=dict(
            orientation="h",
            yanchor="top",
            y=1.1,
            xanchor="center",
            x=0.5
        )
    )

    # Show the figure
    fig.show()

# Call the function to plot total GDP over time for both BRICS versions
plot_total_gdp_bar_chart_with_labels(brics_version="original")
plot_total_gdp_bar_chart_with_labels(brics_version="all_members")


```

## Step 9: Plot the 2023 GDP Pie Chart

This step visualizes the 2023 GDP share for each group (G7, BRICS, and Rest of World) using a pie chart. We use a reusable function, `plot_2023_gdp_pie_chart`, to display either "BRICS - Original" or "BRICS - All Members" based on the specified `brics_version` parameter.


```python
# Define a reusable function to plot the 2023 GDP share pie chart for each group
def plot_2023_gdp_pie_chart(brics_version="original"):
    """
    Plots a pie chart showing the GDP percentage share of G7, BRICS, and Rest of World in 2023.

    Parameters:
    - brics_version (str): Specifies which BRICS grouping to use ("original" or "all_members").
    """
    # Set the BRICS label based on the specified version
    brics_label = "BRICS - All Members" if brics_version == "all_members" else "BRICS - Original"

    # Assign each country to its appropriate group based on the specified BRICS version
    data['Group'] = data.apply(assign_group, axis=1, brics_version=brics_version)

    # Filter data for 2023 and aggregate GDP by group
    aggregated_data_2023 = data[data['Year'] == 2023].groupby('Group').agg({
        'GDP ($Billions)': 'sum'
    }).reset_index()

    # Calculate GDP percentage share for each group
    total_gdp_2023 = aggregated_data_2023['GDP ($Billions)'].sum()
    aggregated_data_2023['GDP Percentage Share (%)'] = (aggregated_data_2023['GDP ($Billions)'] / total_gdp_2023 * 100).round(1)

    # Define custom colors for each group in the pie chart
    colors = {
        brics_label: '#ff5733',   # Red for BRICS
        'G7': '#4b4bff',          # Blue for G7
        'REST OF WORLD': '#32CD32'  # Green for Rest of World
    }

    # Create the pie chart figure
    fig = go.Figure(go.Pie(
        labels=aggregated_data_2023['Group'],
        values=aggregated_data_2023['GDP ($Billions)'],
        textinfo='label+percent+value',
        insidetextorientation='radial',
        texttemplate='%{label}<br>%{percent}<br>$%{value:,.0f}',  # Display label, percent, and value
        marker=dict(colors=[colors[group] for group in aggregated_data_2023['Group']])
    ))

    # Update layout for the pie chart
    fig.update_layout(
        title="2023 GDP Share ($Billions)",
        showlegend=False
    )

    # Show the figure
    fig.show()

# Call the function to plot the 2023 GDP pie chart for both BRICS versions
plot_2023_gdp_pie_chart(brics_version="original")
plot_2023_gdp_pie_chart(brics_version="all_members")

```

## Step 10: Plot the 2023 GDP Pie Charts

In this step, we create two pie charts for the year 2023:

1. **G7 Countries**: Each country in the G7 is represented by a slice of the pie.
2. **BRICS Countries**: The pie chart varies based on the `brics_version` parameter, displaying either "BRICS - Original" or "BRICS - All Members."

Each slice shows the country name, GDP percentage share, and GDP value. Custom colors differentiate each country.


```python
# Define color maps for G7 and BRICS countries
color_map_g7 = {
    'United States': '#FFA500',  # Orange
    'Japan': '#32CD32',          # Lime Green
    'Germany': '#00CED1',        # Dark Turquoise
    'United Kingdom': '#4682B4', # Steel Blue
    'France': '#FF6347',         # Tomato
    'Italy': '#FF8C00',          # Dark Orange
    'Canada': '#20B2AA'          # Light Sea Green
}

color_map_brics_original = {
    'China': '#1E90FF',          # Dodger Blue
    'India': '#FF4500',          # Orange Red
    'Brazil': '#FFD700',         # Gold
    'Russia': '#DC143C',         # Crimson
    'South Africa': '#8A2BE2'    # Blue Violet
}

color_map_brics_all = {
    'China': '#1E90FF',          # Dodger Blue
    'India': '#FF4500',          # Orange Red
    'Brazil': '#FFD700',         # Gold
    'Russia': '#DC143C',         # Crimson
    'South Africa': '#8A2BE2',   # Blue Violet
    'Egypt': '#7FFF00',          # Chartreuse
    'Ethiopia': '#00FA9A',       # Medium Spring Green
    'Iran': '#FF1493',           # Deep Pink
    'Saudi Arabia': '#FF69B4'    # Hot Pink
}

# Map certain country names to a standardized version
country_name_map = {
    'Russian Federation': 'Russia',
    'Egypt, Arab Rep.': 'Egypt',
    'Iran, Islamic Rep.': 'Iran',
    'Saudi Arabia': 'Saudi Arabia'
}

# Define a reusable function to plot the GDP pie charts for 2023
def plot_gdp_pie_chart_2023(brics_version="original"):
    """
    Plots separate pie charts for the GDP share of G7 and BRICS countries in 2023.

    Parameters:
    - brics_version (str): Specifies which BRICS grouping to use ("original" or "all_members").
    """
    # Select BRICS color map and label based on the specified version
    if brics_version == "original":
        brics_countries = BRICS_ORIGINAL
        color_map = color_map_brics_original
        title = "2023 BRICS - Original GDP ($Billions)"
    else:
        brics_countries = BRICS_ALL_MEMBERS
        color_map = color_map_brics_all
        title = "2023 BRICS - All Members GDP ($Billions)"

    # Prepare data for 2023 and standardize country names where necessary
    data_2023 = data[data['Year'] == 2023].copy()
    data_2023['Country Name'] = data_2023['Country Name'].replace(country_name_map)

    # Separate G7 and BRICS data for 2023
    g7_data = data_2023[data_2023['Country Code'].isin(G7_COUNTRIES)].copy()
    brics_data = data_2023[data_2023['Country Code'].isin(brics_countries)].copy()

    # Calculate GDP percentage share for G7 and BRICS countries
    g7_data['GDP Percentage Share'] = (g7_data['GDP ($Billions)'] / g7_data['GDP ($Billions)'].sum() * 100).round(0)
    brics_data['GDP Percentage Share'] = (brics_data['GDP ($Billions)'] / brics_data['GDP ($Billions)'].sum() * 100).round(0)

    # Create G7 pie chart
    fig_g7 = go.Figure(go.Pie(
        labels=g7_data['Country Name'],
        values=g7_data['GDP ($Billions)'],
        text=[f"{country}<br>{int(gdp_share)}%<br>${gdp:,.0f}B" for country, gdp_share, gdp in zip(g7_data['Country Name'], g7_data['GDP Percentage Share'], g7_data['GDP ($Billions)'])],
        textinfo='text',
        marker=dict(colors=[color_map_g7[country] for country in g7_data['Country Name']]),
        hole=0.4  # Donut-style pie chart
    ))
    fig_g7.update_layout(title="2023 G7 GDP ($Billions)", showlegend=False)

    # Create BRICS pie chart
    fig_brics = go.Figure(go.Pie(
        labels=brics_data['Country Name'],
        values=brics_data['GDP ($Billions)'],
        text=[f"{country}<br>{int(gdp_share)}%<br>${gdp:,.0f}B" for country, gdp_share, gdp in zip(brics_data['Country Name'], brics_data['GDP Percentage Share'], brics_data['GDP ($Billions)'])],
        textinfo='text',
        marker=dict(colors=[color_map[country] for country in brics_data['Country Name']]),
        hole=0.4
    ))
    fig_brics.update_layout(title=title, showlegend=False)

    # Display the G7 and BRICS pie charts
    fig_g7.show()
    fig_brics.show()

# Call the function to plot 2023 GDP pie charts for both BRICS versions
plot_gdp_pie_chart_2023(brics_version="original")
plot_gdp_pie_chart_2023(brics_version="all_members")

```

## Step 11: Plot the 2023 GDP Sunburst Chart

In this step, we create a sunburst chart to visualize the GDP distribution for G7 and BRICS countries in 2023. Each group (G7 and BRICS) forms a main branch, with individual countries represented as sub-branches. The function `plot_gdp_sunburst` allows for flexibility, displaying either "BRICS - Original" or "BRICS - All Members" based on the specified `brics_version` parameter.


```python
# Define a reusable function to plot the 2023 GDP sunburst chart
def plot_gdp_sunburst(brics_version="original"):
    """
    Plots a sunburst chart displaying the GDP distribution for G7 and BRICS countries in 2023.

    Parameters:
    - brics_version (str): Specifies which BRICS grouping to use ("original" or "all_members").
    """
    # Define colors for G7 and BRICS groups
    g7_color = '#1f77b4'  # Blue for G7
    brics_color = '#d62728'  # Red for BRICS

    # Create color maps for individual G7 and BRICS countries
    color_map_g7 = {country: g7_color for country in G7_COUNTRIES}
    color_map_brics = {country: brics_color for country in (BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL)}

    # Filter data for the year 2023
    data_2023 = data[data['Year'] == 2023]

    # Define labels, parents, and values for the sunburst chart
    labels = (
        ['G7', 'BRICS'] +
        list(data_2023[data_2023['Country Code'].isin(G7_COUNTRIES)]['Country Name']) +
        list(data_2023[data_2023['Country Code'].isin(BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL)]['Country Name'])
    )

    parents = (
        ['', ''] +
        ['G7'] * len(G7_COUNTRIES) +
        ['BRICS'] * (len(BRICS_ALL_MEMBERS) if brics_version == "all_members" else len(BRICS_ORIGINAL))
    )

    # Sum GDP for G7 and BRICS, then add individual country GDP values
    values = (
        [data_2023[data_2023['Country Code'].isin(G7_COUNTRIES)]['GDP ($Billions)'].sum(),
         data_2023[data_2023['Country Code'].isin(BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL)]['GDP ($Billions)'].sum()] +
        list(data_2023[data_2023['Country Code'].isin(G7_COUNTRIES)]['GDP ($Billions)']) +
        list(data_2023[data_2023['Country Code'].isin(BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL)]['GDP ($Billions)'])
    )

    # Assign colors for each group and country
    colors = (
        [g7_color, brics_color] +
        [color_map_g7[country] for country in G7_COUNTRIES] +
        [color_map_brics[country] for country in (BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL)]
    )

    # Create the sunburst chart
    fig = go.Figure(go.Sunburst(
        labels=labels,
        parents=parents,
        values=values,
        marker=dict(colors=colors),
        branchvalues="total",
        textinfo="label+value"  # Display both label and GDP value
    ))

    # Update layout for the sunburst chart
    fig.update_layout(
        title_text=f"2023 GDP ($Billions) for {'BRICS - All Members' if brics_version == 'all_members' else 'BRICS - Original'}",
        title_x=0.5,
        margin=dict(t=50, l=0, r=0, b=0)
    )

    # Display the sunburst chart
    fig.show()

# Call the function to plot the sunburst chart for both BRICS versions
plot_gdp_sunburst(brics_version="original")
plot_gdp_sunburst(brics_version="all_members")


```

## Step 12: Plot the 2023 GDP Treemap Chart

In this step, we create a treemap chart to visualize the GDP distribution among G7, BRICS, and Rest of World countries in 2023. Each group (G7, BRICS, and Rest of World) is represented by a primary block, with individual countries displayed as smaller blocks within each group. The function `plot_gdp_treemap` allows for flexibility, displaying either "BRICS - Original" or "BRICS - All Members" based on the `brics_version` parameter.


```python
# Define a reusable function to plot the 2023 GDP treemap chart
def plot_gdp_treemap(brics_version="original"):
    """
    Plots a treemap chart displaying the GDP distribution for G7, BRICS, and Rest of World countries in 2023.

    Parameters:
    - brics_version (str): Specifies which BRICS grouping to use ("original" or "all_members").
    """
    # Define colors for G7, BRICS, and Rest of World groups
    g7_color = '#1f77b4'         # Blue for G7
    brics_color = '#d62728'      # Red for BRICS
    rest_of_world_color = '#add8e6'  # Light blue for Rest of World

    # Create color maps for individual G7 and BRICS countries
    color_map_g7 = {country: g7_color for country in G7_COUNTRIES}
    color_map_brics = {country: brics_color for country in (BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL)}

    # Prepare data for the year 2023
    data_2023 = data[data['Year'] == 2023].copy()
    data_2023['GDP ($Billions)'] = data_2023['GDP ($Billions)'].fillna(0)  # Handle any missing GDP values

    # Define labels and values for the treemap chart
    labels = (
        ['G7', 'BRICS', 'Rest of World'] +
        [f"{country} ${int(gdp):,}" for country, gdp in zip(data_2023['Country Name'], data_2023['GDP ($Billions)'])]
    )

    parents = (
        ['', '', ''] +
        ['G7' if code in G7_COUNTRIES else 'BRICS' if code in (BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL) else 'Rest of World'
         for code in data_2023['Country Code']]
    )

    # Aggregate GDP for G7, BRICS, and Rest of World groups
    values = (
        [int(data_2023[data_2023['Country Code'].isin(G7_COUNTRIES)]['GDP ($Billions)'].sum()),
         int(data_2023[data_2023['Country Code'].isin(BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL)]['GDP ($Billions)'].sum()),
         int(data_2023[~data_2023['Country Code'].isin(G7_COUNTRIES + (BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL))]['GDP ($Billions)'].sum())
        ] + [int(gdp) for gdp in data_2023['GDP ($Billions)']]
    )

    # Assign colors to each country based on group
    colors = (
        [g7_color, brics_color, rest_of_world_color] +
        [color_map_g7[code] if code in G7_COUNTRIES else color_map_brics[code] if code in (BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL) else rest_of_world_color
         for code in data_2023['Country Code']]
    )

    # Create the treemap chart
    fig = go.Figure(go.Treemap(
        labels=labels,
        parents=parents,
        values=values,
        marker=dict(colors=colors),
        branchvalues="total",
        textinfo="label"  # Display country name and GDP value
    ))

    # Update layout for the treemap chart
    fig.update_layout(
        title_text=f"2023 GDP Treemap Chart for {'BRICS - All Members' if brics_version == 'all_members' else 'BRICS - Original'}",
        title_x=0.5,
        margin=dict(t=50, l=0, r=0, b=0)
    )

    # Display the treemap chart
    fig.show()

# Call the function to plot the treemap chart for both BRICS versions
plot_gdp_treemap(brics_version="original")
plot_gdp_treemap(brics_version="all_members")


```

## Step 13 - Plot the Historic Trend of GDP Per Capita

## Step 13: Plot the Historic Trend of GDP Per Capita

This step visualizes the historic trend of GDP per capita for each group (G7, BRICS, and Rest of World) over time. The function `plot_gdp_per_capita_trend` allows us to select either "BRICS - Original" or "BRICS - All Members" based on the `brics_version` parameter. Each group is represented by a distinct color.


```python
# Define a reusable function to plot the historic trend of GDP per capita
def plot_gdp_per_capita_trend(brics_version="original"):
    """
    Plots the historic trend of GDP per capita for G7, BRICS, and Rest of World groups.

    Parameters:
    - brics_version (str): Specifies which BRICS grouping to use ("original" or "all_members").
    """
    # Check if necessary columns are present in the dataset
    required_columns = {'Year', 'Country Code', 'GDP ($Billions)', 'Population (Millions)'}
    if not required_columns.issubset(data.columns):
        print("One or more necessary columns are missing in the dataset.")
        return

    # Calculate GDP per capita
    data['GDP Per Capita ($)'] = (data['GDP ($Billions)'] * 1e9) / (data['Population (Millions)'] * 1e6)

    # Define the BRICS countries based on the specified version
    brics_countries = BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL

    # Assign each country to its appropriate group
    data['Group'] = data['Country Code'].apply(
        lambda code: 'G7' if code in G7_COUNTRIES else 'BRICS' if code in brics_countries else 'Rest of World'
    )

    # Group by year and group, then calculate the mean GDP per capita for each group
    gdp_per_capita_grouped = data.groupby(['Year', 'Group']).agg({'GDP Per Capita ($)': 'mean'}).reset_index()
    grouped_pivot = gdp_per_capita_grouped.pivot(index='Year', columns='Group', values='GDP Per Capita ($)').fillna(0)

    # Define colors for each group
    color_map = {'BRICS': '#ff5733', 'G7': '#4b4bff', 'Rest of World': '#32CD32'}

    # Create the figure
    fig = go.Figure()

    # Add each group as a bar trace in the stacked bar chart
    for group in ['BRICS', 'G7', 'Rest of World']:
        fig.add_trace(go.Bar(
            x=grouped_pivot.index,
            y=grouped_pivot[group],
            name=group,
            marker=dict(color=color_map[group]),
            text=[f"${value:,.0f}" for value in grouped_pivot[group]],  # Display GDP per capita values
            textposition="inside"
        ))

    # Update layout for the chart
    fig.update_layout(
        title=f"GDP Per Capita Historic Trend ({'BRICS All Members' if brics_version == 'all_members' else 'BRICS Original'})",
        xaxis=dict(title="Year"),
        yaxis=dict(title="GDP Per Capita ($)", tickprefix="$"),
        barmode='stack',  # Stack bars by year
        legend=dict(title='', orientation="h", y=1.1, x=0.5, xanchor="center"),
        paper_bgcolor='white',
        plot_bgcolor='white'
    )

    # Show the figure
    fig.show()

# Call the function to plot GDP per capita trends for both BRICS versions
plot_gdp_per_capita_trend(brics_version="original")
plot_gdp_per_capita_trend(brics_version="all_members")


```

```python
# Define a reusable function to plot the 2023 GDP per capita bar chart for each group
def plot_gdp_per_capita_bar_chart(brics_version="original"):
    """
    Plots a bar chart displaying the 2023 GDP per capita for each group (G7, BRICS, and Rest of World).

    Parameters:
    - brics_version (str): Specifies which BRICS grouping to use ("original" or "all_members").
    """
    # Check if necessary columns are present in the dataset
    required_columns = {'Year', 'Country Code', 'GDP ($Billions)', 'Population (Millions)'}
    if not required_columns.issubset(data.columns):
        print("One or more necessary columns are missing in the dataset.")
        return

    # Filter data for the year 2023 and calculate GDP per capita
    data_2023 = data[data['Year'] == 2023].copy()
    data_2023['GDP Per Capita ($)'] = (data_2023['GDP ($Billions)'] * 1e9) / (data_2023['Population (Millions)'] * 1e6)

    # Define BRICS countries based on the specified version
    brics_countries = BRICS_ALL_MEMBERS if brics_version == "all_members" else BRICS_ORIGINAL

    # Assign each country to its appropriate group
    data_2023['Group'] = data_2023['Country Code'].apply(
        lambda code: 'G7' if code in G7_COUNTRIES else 'BRICS' if code in brics_countries else 'Rest of World'
    )

    # Group by 'Group' and calculate the mean GDP per capita for each group
    gdp_per_capita_grouped = data_2023.groupby('Group').agg({'GDP Per Capita ($)': 'mean'}).reset_index()
    gdp_per_capita_grouped['GDP Per Capita ($)'] = gdp_per_capita_grouped['GDP Per Capita ($)'].round(0)

    # Sort groups by GDP per capita
    gdp_per_capita_grouped = gdp_per_capita_grouped.sort_values(by='GDP Per Capita ($)', ascending=False)

    # Define colors for each group
    color_map = {'BRICS': '#ff5733', 'G7': '#4b4bff', 'Rest of World': '#32CD32'}

    # Plot the bar chart
    fig = go.Figure()
    for _, row in gdp_per_capita_grouped.iterrows():
        fig.add_trace(go.Bar(
            x=[row['Group']],
            y=[row['GDP Per Capita ($)']],
            name=row['Group'],
            marker_color=color_map[row['Group']],
            text=f"${int(row['GDP Per Capita ($)']):,}",  # Display GDP per capita value
            textposition="inside"
        ))

    # Set the title based on BRICS version
    title = "2023 GDP Per Capita ($) - " + ("BRICS All Members" if brics_version == "all_members" else "BRICS Original")

    # Update layout for the bar chart
    fig.update_layout(
        title=title,
        xaxis=dict(title="Group"),
        yaxis=dict(title="GDP Per Capita ($)", tickprefix="$"),
        barmode='group',
        paper_bgcolor='white',
        plot_bgcolor='white',
        showlegend=False
    )

    # Show the figure
    fig.show()

# Call the function to plot GDP per capita for both BRICS versions
plot_gdp_per_capita_bar_chart(brics_version="original")
plot_gdp_per_capita_bar_chart(brics_version="all_members")

```

## Step 16 - Create a Choropleth

## Step 16: Create a Choropleth Map

In this step, we create a choropleth map to display the G7 and BRICS countries. The map colors G7 and BRICS countries distinctly and annotates each country with its name. The function `plot_choropleth_map` allows us to select either "BRICS - Original" or "BRICS - All Members" based on the `brics_version` parameter.


```python
import plotly.express as px
import pandas as pd

# Data for G7 and BRICS countries
data = pd.DataFrame({
    'Country': ['United States', 'Canada', 'Germany', 'France', 'United Kingdom', 'Italy', 'Japan',
                'China', 'India', 'Brazil', 'South Africa', 'Russia', 'Saudi Arabia', 'Iran', 'Egypt', 'Ethiopia'],
    'Group': ['G7', 'G7', 'G7', 'G7', 'G7', 'G7', 'G7',
              'BRICS', 'BRICS', 'BRICS', 'BRICS', 'BRICS', 'BRICS', 'BRICS', 'BRICS', 'BRICS']
})

# Define BRICS and G7 country groups
BRICS_ORIGINAL = ['China', 'India', 'Brazil', 'South Africa', 'Russia']
BRICS_ALL_MEMBERS = BRICS_ORIGINAL + ['Saudi Arabia', 'Iran', 'Egypt', 'Ethiopia']
G7_COUNTRIES = ['United States', 'Canada', 'Germany', 'France', 'United Kingdom', 'Italy', 'Japan']

# Define a reusable function to plot the G7 and BRICS choropleth map
def plot_choropleth_map(brics_version="original"):
    """
    Plots a choropleth map highlighting G7 and BRICS countries.

    Parameters:
    - brics_version (str): Specifies which BRICS grouping to use ("original" or "all_members").
    """
    # Select BRICS members based on version
    if brics_version == "all_members":
        brics_countries = BRICS_ALL_MEMBERS
        title = "Choropleth Map of G7 and BRICS All Members"
    else:
        brics_countries = BRICS_ORIGINAL
        title = "Choropleth Map of G7 and BRICS Original"

    # Filter data to include only G7 and BRICS countries
    data['Group'] = data['Country'].apply(
        lambda x: 'G7' if x in G7_COUNTRIES else 'BRICS' if x in brics_countries else None
    )
    filtered_data = data.dropna()  # Drop rows where Group is None (non-G7 and non-BRICS countries)

    # Define colors for each group
    color_map = {'G7': '#4b4bff', 'BRICS': '#ff5733'}

    # Create the Plotly choropleth map
    fig = px.choropleth(
        filtered_data,
        locations="Country",
        locationmode="country names",
        color="Group",
        color_discrete_map=color_map,
        title=title,
        hover_name="Country"
    )


    # Lock the projection scale and remove x and y axis
    fig.update_geos(
        showframe=False,
        showcoastlines=True,
        coastlinecolor="black",
        projection_type="equirectangular",
        landcolor="white",
        showlakes=False,
        countrycolor="black",
    )

    # Add annotations for country names with precise coordinates
    annotations = {
        'United States': (-70, 38), 'Canada': (-80, 50), 'Germany': (10, 51),
        'France': (2, 46), 'United Kingdom': (-1, 54), 'Italy': (12, 43), 'Japan': (138, 36),
        'China': (104, 35), 'India': (78, 22), 'Brazil': (-53, -10), 'South Africa': (24, -30),
        'Russia': (105, 60), 'Saudi Arabia': (45, 23), 'Iran': (53, 32),
        'Egypt': (30, 27), 'Ethiopia': (39, 9)
    }

    for country, (lon, lat) in annotations.items():
        if country in brics_countries or country in G7_COUNTRIES:
            fig.add_annotation(
                x=lon, y=lat,
                xref="x", yref="y",
                text=country,
                showarrow=False,
                font=dict(size=9, color="black"),
                align="center"
            )

    # Set fixed layout dimensions and remove visible x/y axes
    fig.update_layout(
        autosize=False,
        width=800,
        height=450,
        xaxis_visible=False,
        yaxis_visible=False,
        paper_bgcolor='white',
        plot_bgcolor='white',
        legend_title_text='Group',
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=-0.2,
            xanchor="center",
            x=0.5,
            bordercolor="black",
            borderwidth=1
        )
    )

    # Show the figure
    fig.show()

# Plot with both versions
plot_choropleth_map(brics_version="original")
plot_choropleth_map(brics_version="all_members")

```

```python
COLOR_MAP = {
    "REST OF WORLD" : "lightgreen",
    "BRICS" : "orangered",
    "G7" : "MediumSlateBlue"
}


def geography(df):
    df_2020 = df[(df["Group"].isin(["G7", "BRICS"])) & (df["Year"] == 2020)]
    fig = px.choropleth(
        df_2020,
        locations='Country Code',
        color='Group',
        scope="world",
        labels="Country Name",
        color_discrete_map=COLOR_MAP,
        featureidkey='properties.name',
    )
    ## Add labels on countries
    fig.add_scattergeo(
        locations=df_2020["Country Code"],
        text=df_2020['Country Name'],
        mode='text',
        textfont_color="black",
        featureidkey='properties.name',
        showlegend=False
    )
    fig.update_geos(
        fitbounds='locations',
        oceancolor='blue'
    )
    fig.update_layout(legend=dict(
        yanchor="bottom",
        y=0.05,
        xanchor="center",
        x=0.5,
        orientation="h",
        title=None
    ))
    fig.update_layout(margin=dict(l=0.5, r=0.5, t=0.5, b=0.5))
    return fig

```

```python
url = 'https://raw.githubusercontent.com/mineozcan/UMBC-DATA609-FALL24-WANG/refs/heads/main/2024-11-11T02-35_export.csv'

data = pd.read_csv(url) # Load the dataset into a pandas DataFrame

fig = geography(data)

fig.show()
```

```python
def country_bar(df, column, year, sign=None):
    fig = px.bar(
        df[df["Group"].isin(["G7", "BRICS"]) & (df["Year"] == year)],
        x="Country Name",
        y=column,
        color="Group",
        text=column,
        color_discrete_map=COLOR_MAP,
        title=f"{year} {column}"
    )

    fig.update_layout(
        showlegend=True,
        xaxis={'categoryorder': 'total descending'}
    )

    if sign == "$":
        temp ='$%{text:,.0f}'
    elif sign == "%":
        temp ='%{text}%'
    else:
        temp = '%{text}'

    fig.update_traces(
        texttemplate=temp,
        textposition='outside'
    )

    fig.update_layout(
        legend=dict(
            orientation="h",  # Horizontal legend
            x=1,              # at the end of x
            xanchor="right",  # Anchor the legend's right
            yanchor="bottom", # Anchor the legend's bottom
            y=1.05,           # Slightly above the plot
            title_text=None
        )
    )

    return fig
```
