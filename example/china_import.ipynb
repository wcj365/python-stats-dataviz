```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_excel("https://github.com/wcj365/python-stats-dataviz/raw/refs/heads/master/example/Imports_from_Counterpart_Countries.xlsx", skiprows=6)
df.info()
```

```python
df.head(1).T
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
df.head()
```

```python
df.rename(columns={"Unnamed: 1" : "Country"}, inplace=True)
df.head()
```

```python
print(list(df["Country"]))
```

```python
G7 = [
    "Japan",
    "United States",
    "Canada",
    "United Kingdom",
    "Germany",
    "France",
    "Italy"
]
```

```python
BRICS = [
    "Brazil",
    "Russian Federation",
    "India",
    "China",
    "South Africa"
]
```

```python
df_brics = df[df["Country"].isin(BRICS)]
df_brics
```

```python
df_g7 = df[df["Country"].isin(G7)]
df_g7
```

```python
def assign_group(country):
    if country in BRICS:
        return "BRICS"
    elif country in G7:
        return "G7"
    else:
        return "REST OF WORLD"


df["Group"] = df["Country"].apply(assign_group)

df.sample(5)
```

```python
df["Group"].value_counts()
```

```python
df_group = df.select_dtypes(include='number').groupby(df["Group"]).sum().reset_index()
df_group
```

```python
df_melt =pd.melt(df_group, id_vars=["Group"], var_name='Year', value_name='Export')
df_melt
```

```python
fig = px.bar(df_melt, x="Year", y="Export", color="Group", barmode="group")
fig.show()
```

```python
# Compute the total value per category

df_total = df_melt.groupby("Year")['Export'].transform('sum')

df_total
```

```python
# Compute the percentage of the total value within each group
df_melt['Percentage'] = round((df_melt['Export'] / df_total) * 100, 1)

df_melt
```

```python
fig = px.bar(df_melt, x="Year", y="Percentage", color="Group", barmode="stack")
fig.show()
```
